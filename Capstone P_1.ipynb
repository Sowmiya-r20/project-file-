```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1: Probability of scoring given the player is on the field

# Define scoring probabilities
data['Scoring_Probability'] = data['goals'] / data['minutes_played']

# Top players by scoring probability
top_prob_players = data[['player_name_y', 'Scoring_Probability']].sort_values(by='Scoring_Probability', ascending=False).head(10)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Scoring_Probability', y='player_name_y', data=top_prob_players, palette='coolwarm')
plt.title('Top Players by Scoring Probability')
plt.xlabel('Scoring Probability (Goals per Minute)')
plt.ylabel('Player Name')
plt.show()

```

#### Interpretation:
This shows the likelihood of a player scoring a goal while on the field, providing insight into their efficiency.

```python
#2:Conditional probability of a team winning given a specific manager
# Using Conditional probability
if 'home_club_goals' in df.columns and 'away_club_goals' in df.columns and 'goals' in df.columns:
    # Total matches
    total_matches = len(df)
    
    # Calculate home wins
    home_wins = len(df[df['home_club_goals'] > df['away_club_goals']])
    prob_home_win = home_wins / total_matches
    
    # Calculate matches where goals were scored
    matches_with_goals = len(df[df['goals'] > 0])
    
    # Calculate matches where the home team won and goals were scored
    home_wins_with_goals = len(df[(df['home_club_goals'] > df['away_club_goals']) & (df['goals'] > 0)])
    prob_home_win_given_goals = home_wins_with_goals / matches_with_goals if matches_with_goals > 0 else 0
    
    # Print results
    print(f"Probability of a Home Win: {prob_home_win:.2f}")
    print(f"Conditional Probability of Home Win Given Goals Scored: {prob_home_win_given_goals:.2f}")
else:
    print("Columns 'home_club_goals', 'away_club_goals', or 'goals' are missing.")
```

#### Interpretation:
This calculates the probability of a home team winning, given a specific manager. High probabilities indicate strong leadership.

```python
#3: 
# Aggregate the goals scored by home and away teams
home_performance = data.groupby('home_club_goals')['home_club_manager_name'].sum().reset_index()
home_performance = home_performance.rename(columns={'home_club_goals': 'Team', 'home_club_manager_name': 'Total Home Goals'})

away_performance = data.groupby('away_club_goals')['away_club_manager_name'].sum().reset_index()
away_performance = away_performance.rename(columns={'away_club_goals': 'Team', 'away_club_manager_name': 'Total Away Goals'})

# Sort and display the top 5 performing teams for home and away
top_home_teams = home_performance.sort_values(by='Total Home Goals', ascending=False).head(5)
top_away_teams = away_performance.sort_values(by='Total Away Goals', ascending=False).head(5)

# Display the results
print("Top 5 Home Teams by Goals Scored:")
print(top_home_teams)

print("Top 5 Away Teams by Goals Scored:")
print(top_away_teams)
```

#### Interpretation:
This analysis highlights the most successful teams in terms of goal-scoring performance both at home and away.

```python
# Normality of Goals Distribution across matches 
plt.figure(figsize=(8, 6))
sns.histplot(df['goals'], bins=5, kde=True, color='gold')
plt.title('Distribution of Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.show()
```

#### Interpretation:
By observing the plot, you can derive specific conclusions about the goals' distribution, such as whether they follow a typical pattern or have irregularities that may need further investigation..

```python
# What does the Central Limit Theorem (CLT) reveal about the distribution of sample means for the "goals" column in the dataset?
if 'goals' in data.columns:
    sample_means = []
    for _ in range(1000):
        sample = df['goals'].dropna().sample(50, replace=True)
        sample_means.append(sample.mean())
    
    # Visualization
    plt.figure(figsize=(10, 6))
    sns.histplot(sample_means, kde=True, color='Green')
    plt.title('Distribution of Sample Means (CLT)', fontsize=16)
    plt.xlabel('Sample Mean of Goals', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
else:
    print("Column 'goals' is missing.")
```

Interpretation: 
The CLT states that the distribution of sample means approaches a normal distribution, regardless of the original distribution of the "goals" column, as long as the sample size is sufficiently large. The histogram and KDE plot of the sample means visually confirm this, showing a bell-shaped curve centered around the true mean of the "goals" column.
